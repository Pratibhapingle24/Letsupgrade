```
def isPrime(n):
    if n > 1:
        for i in range(2,n):
            if n % i ==0:
                return False
        else:
            return True
```

```
isPrime(11)
```

```
isPrime(4)
```

```
primeno=[]
```

```
lst = list(range(0,2500))
```

```
for item in lst:
    if isPrime(item):
        primeno.append(item)
```

```
print(primeno)
```

```
lst_prime_no = filter(isPrime,lst)
```
